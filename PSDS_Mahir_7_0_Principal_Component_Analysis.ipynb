```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#inputkan data
A = np.matrix([[1,2,3,4],[5,5,6,7],[1,4,2,3],[5,3,2,1],[8,1,2,2]])
```

```python
df = pd.DataFrame(A,columns=['a1','a2','a3','a4'])
df
```

```python
df.mean()
```

```python
std_dev = np.std(df)
std_dev
```

```python
# Menginstansiasikan scaler
scaler = StandardScaler()

# Melatih scaler pada dataset
scaler.fit(df)

# Mentransformasi dataset
df_scaled = scaler.transform(df)

print(df_scaled)
```

```python
dff = pd.DataFrame(df_scaled,columns=['a1','a2','a3','a4'])
dff
```

```python
#mean data yang sudah dinormalkan
dff.mean()
```

```python
#mencari varians yang dijadikan diagonal
var = np.var(dff)
var
```

```python
cov = np.cov(dff.T, ddof=0) #ddf digunakan untuk menghitung matriks dengan derajat kebebasan 0
cov
```

```python
dff.T
```

```python
#mencari nilai eigen dan vektor eigen
eigenvalues, eigenvectors = np.linalg.eig(cov)
print("Nilai Eigen:")
print(eigenvalues)
print("\nVektor Eigen:")
print(eigenvectors)
```

```python
#Transformasikan kembali matriks dengan cara mengalikan matriks dengan vektor eigen dari lamda yang dipilih

import numpy as np
a = df_scaled
b = np.matrix([
    [0.16195986, -0.91705888],
    [-0.52404813, 0.20692161],
    [-0.58589647, -0.3205394],
    [-0.59654663, -0.11593512]])
c = np.dot(a,b)
c
```

```python
#mencari PCA secara langsung
pca = PCA(n_components=2)
pr = pca.fit_transform(dff)
prf = pd.DataFrame(data=pr,columns=["a1","a2"])
prf
```
